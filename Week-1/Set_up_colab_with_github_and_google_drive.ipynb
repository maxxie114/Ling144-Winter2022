## Step 2: Change your working directory

Google Colabs allow you to execute python commands as well as shell commands. We will generally be using Colabs to execute Python code, which is how you mounted the Google Drive above. ## Step 3: Generate a Github Access Token

You now need to link your Google Drive account to your Github account so that you can use Google Colabs as an intermediary to sync files from one location to another.

1. Log in to your Github account.
2. Click on your avatar in the upper right of the screen and navigate to **Settings > Developer Settings > Personal Access Tokens**
3. Click on the **Generate New Token** button in the top right
4. In the form that appears, fill in the following:
   * **Note**: Github Access token for Colabs
   * **Expiration**: 90 days
   * **Select Scopes**: 
      * *check the* **repo** *checkbox*
      * *check the* **admin:repo_hook** *checkbox*
      * *check the* **admin:public_key** *checkbox*
5. Scroll to the bottom of the page and click the **Generate token** button
6. Copy the token that appears in the green box and paste somewhere accessible.
